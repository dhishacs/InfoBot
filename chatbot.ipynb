Import Libraries

```python
import numpy as np
import pandas as pd
import re
import torch
import random
import torch.nn as nn
import transformers
import matplotlib.pyplot as plt
```

Load Dataset

```python
# We have prepared a chitchat dataset with 5 labels
df = pd.read_excel("/content/excel with the question and its intents.xlsx")
df.head()
```

```python
df['label'].value_counts()
```

```python
# Converting the labels into encodings
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])
# check class distribution
df['label'].value_counts(normalize = True)
```

```python
train_text, train_labels = df['text'], df['label']
```

Data Preparation

BERT Model

```python
from transformers import AutoModel, BertTokenizerFast

# Load the BERT tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')


# Import BERT-base pretrained model
bert = AutoModel.from_pretrained('bert-base-uncased')
```

Sample data for distilbert-base-uncased tokenizer

```python
#EXAMPLE ENCODING
text = ["This is our chatbot project", "We created this for AMCS department"]

encoded_input = tokenizer(text, padding=True,truncation=True, return_tensors='pt')
input_ids = encoded_input['input_ids']
attention_mask = encoded_input['attention_mask']

print("Input IDs:", input_ids)
print("Attention Mask:", attention_mask)
'''In input_ids:
101 - Indicates beginning of the sentence
102 - Indicates end of the sentence
In attention_mask:
1 - Actual token
0 - Padded token'''
```

```python
# tokenize and encode sequences in the training set
tokens_train = tokenizer(
    train_text.tolist(),
    max_length = 13,
    padding = True,
    truncation=True,
    return_token_type_ids=False
)
```

Next, we will convert the integer sequences to tensors.

```python
# for train set
train_seq = torch.tensor(tokens_train['input_ids'])
train_mask = torch.tensor(tokens_train['attention_mask'])
train_y = torch.tensor(train_labels.tolist())
```

Now we will create dataloaders for the training set. These dataloaders will pass batches of train data as input to the model during the training phase.

```python
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler
#define a batch size
batch_size = 10

# wrap tensors
train_data = TensorDataset(train_seq, train_mask, train_y)

# sampler for sampling the data during training
train_sampler = RandomSampler(train_data)

# DataLoader for train set
train_dataloader = DataLoader(train_data, sampler=train_sampler, batch_size=batch_size)
```

Define Model Architecture

```python
class BERT_Arch(nn.Module):
   def __init__(self, bert):
       super(BERT_Arch, self).__init__()
       self.bert = bert

       # dropout layer
       self.dropout = nn.Dropout(0.2)

       # relu activation function
       self.relu =  nn.ReLU()
       # dense layer
       self.fc1 = nn.Linear(768,512)
       self.fc2 = nn.Linear(512,256)
       self.fc3 = nn.Linear(256,14)

       #softmax activation function
       self.softmax = nn.LogSoftmax(dim=1)

    #define the forward pass
   def forward(self, sent_id, mask):
      #pass the inputs to the model
      cls_hs = self.bert(sent_id, attention_mask=mask)[0][:,0]

      x = self.fc1(cls_hs)
      x = self.relu(x)
      x = self.dropout(x)

      x = self.fc2(x)
      x = self.relu(x)
      x = self.dropout(x)
      # output layer
      x = self.fc3(x)

      # apply softmax activation
      x = self.softmax(x)
      return x
```

```python
# freeze all the parameters. This will prevent updating of model weights during fine-tuning.
for param in bert.parameters():
      param.requires_grad = False
model = BERT_Arch(bert)
model = model.to('cpu')
from torchinfo import summary
summary(model)
```

Fine-tuning of the model

Optimizer

```python
import torch.optim as optim
optimizer = optim.AdamW(model.parameters(), lr=1e-5)
```

Find Class Weights

```python
from sklearn.utils.class_weight import compute_class_weight
#compute the class weights
class_wts = compute_class_weight(class_weight = "balanced",
                                        classes = np.unique(train_labels),
                                        y = train_labels)
print(class_wts)
```

Balancing the weights while calculating the error

```python
# convert class weights to tensor
weights= torch.tensor(class_wts,dtype=torch.float)
weights = weights.to('cpu')
# loss function
cross_entropy = nn.NLLLoss(weight=weights)
```

Setting up the epochs

```python
# empty lists to store training and validation loss of each epoch
train_losses=[]
# number of training epochs
epochs = 400
```

Fine-Tune the model

```python
import numpy as np

def train():
    model.train()
    total_loss = 0
    total_correct = 0
    total_samples = 0
    total_preds = []

    for step, batch in enumerate(train_dataloader):
        batch = [r.to('cpu') for r in batch]
        sent_id, mask, labels = batch
        preds = model(sent_id, mask)
        loss = cross_entropy(preds, labels)
        total_loss += loss.item()

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        preds = preds.detach().cpu().numpy()
        total_preds.append(preds)

        # Compute the number of correct predictions in the current batch
        correct_preds = np.sum(np.argmax(preds, axis=1) == labels.cpu().numpy())
        total_correct += correct_preds
        total_samples += len(labels)

    avg_loss = total_loss / len(train_dataloader)

    total_preds = np.concatenate(total_preds, axis=0)

    # Calculate training accuracy
    training_accuracy = (total_correct / total_samples) * 100.0

    return avg_loss,training_accuracy,total_preds
```

Start Model Training

```python
for epoch in range(epochs):

    print('\n Epoch {:} / {:}'.format(epoch + 1, epochs))

    #train model
    train_loss,accuracy, _ = train()

    # append training and validation loss
    train_losses.append(train_loss)
    # it can make your experiment reproducible, similar to set  random seed to all options where there needs a random seed.
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
print(f'\nTraining Loss: {train_loss:.3f}')
print(f"Training Accuracy: {accuracy:.2f}%")
```

```python
%%writefile App.py

import streamlit as st
st.set_page_config(page_title='InfoBot', layout="wide")

st.title('InfoBot')
st.markdown('---')


import streamlit as st
from transformers import BertForQuestionAnswering, BertTokenizer

st.title("BERT Chatbot")

@st.cache(allow_output_mutation=True)
def load_model():
    tokenizer = BertTokenizer.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")
    model = BertForQuestionAnswering.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")
    return model, tokenizer

model, tokenizer = load_model()

user_question = st.text_input("Ask a question:")

if user_question:
    inputs = tokenizer(user_question, "Context: BERT is great for chatbots.", add_special_tokens=True, return_tensors="pt")
    start_scores, end_scores = model(**inputs)
    answer_start = torch.argmax(start_scores)
    answer_end = torch.argmax(end_scores)
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs.input_ids[0][answer_start:answer_end + 1]))
    st.write("InfoBot:", answer)


st.sidebar.write('Developed by ')
st.sidebar.write('C S Dhisha - 21PD05')
st.sidebar.write('V K Deeksha - 21PD37')
```
